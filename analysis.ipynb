```python
# Group by "Age" and calculate the mean "Fatality"
df_Age_Fatality = df[["Age","Fatality"]].groupby("Age").agg({"Fatality":"mean"})

#Dictionary
age_groups = {range(0, 16): "0-15", 
              range(16, 31): "16-30", 
              range(31, 46): "31-45", 
              range(46, 61): "46-60", 
              range(61, 81): "61-80", 
              range(81, 150): "+81"}

# Mapping
def ages_groups(age, groups):
    for i, group in groups.items():
        if age in i:
            return group
    return None

df_Age_Fatality['Age_Group'] = df_Age_Fatality.index.map(lambda age: ages_groups(age, age_groups))

df_Age_Fatality = df_Age_Fatality.groupby("Age_Group").agg({"Fatality":"mean"})
df_Age_Fatality 
```
